```sql
CREATE TABLE employee_salary (
emp_id INT PRIMARY KEY,
emp_name VARCHAR(50),
base_salary DECIMAL(10,2),
bonus DECIMAL(10,2),
tax_percent DECIMAL(5,2),
experience_years INT
);

INSERT INTO employee_salary VALUES
(1, 'Karthik', 75000.75, 5000.50, 10.00, 6),
(2, 'Veena', 65000.40, 4000.25, 8.50, 4),
(3, 'Ravi', 85000.90, 6000.75, 12.00, 8),
(4, 'Anil', 70000.10, NULL, 9.00, 5),
(5, 'Suresh', 60000.55, 3000.30, 7.50, 3);
```

```sql
Select emp_name , ABS(base_salary  - coalesce(bonus,0)) as total_salary from employee_salary 
```

```sql
Select Round(base_salary , 0) as rounded_salary from employee_salary 
```

```sql
select
emp_name,
ROUND(base_salary,1) AS rounded_salary
FROM employee_salary;
```

```sql

SELECT 
emp_name,
base_salary,
CEIL(base_salary) AS rounded_up_salary
FROM employee_salary;
     
```

```sql
SELECT 
emp_name,
FLOOR(base_salary) AS rounded_down_salary
FROM employee_salary;
```

```sql
SELECT 
emp_name,
floor(base_salary,0) AS truncated_salary
FROM employee_salary;
```

```sql

SELECT 
emp_name,
experience_years,
MOD(experience_years,2) AS remainder
FROM employee_salary;
```

```sql

SELECT 
emp_name,
POWER(experience_years,2) AS experience_square
FROM employee_salary;
```

```sql

SELECT 
emp_name,
SQRT(base_salary) AS sqrt_salary
FROM employee_salary;
```

```sql
SELECT 
emp_name,
SIGN(base_salary) AS salary_sign
FROM employee_salary;
```

```sql

SELECT 
emp_name,
format_number(base_salary,2) AS formatted_salary
FROM employee_salary;
```

```sql

SELECT 
emp_name,
GREATEST(base_salary, COALESCE(bonus,0)) AS highest_component
FROM employee_salary;
```

```sql
SELECT 
emp_name,
LEAST(base_salary, COALESCE(bonus,0)) AS lowest_component
FROM employee_salary;
```

```sql

SELECT 
emp_name,
base_salary - (base_salary * tax_percent/100) AS salary_after_tax
FROM employee_salary;
```

```sql

SELECT 
emp_name,
base_salary + COALESCE(bonus,0) AS total_compensation
FROM employee_salary;
```

```sql
SELECT 
emp_name,
ROUND(base_salary * tax_percent/100,2) AS tax_amount
FROM employee_salary;
     
```

```sql
SELECT MAX(base_salary) AS max_salary
FROM employee_salary;
```

```sql
SELECT 
emp_name,
FLOOR(base_salary/10000) AS salary_band
FROM employee_salary;
     
```

```sql
SELECT 
emp_name,
ROUND(RAND()*9+1) AS performance_score
FROM employee_salary;
```
