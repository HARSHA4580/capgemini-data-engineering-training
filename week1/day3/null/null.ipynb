```sql
SELECT emp_id, coalesce(salary,0) AS salary
FROM Employees;
```

```sql
select emp_id,coalesce(bonus,0) as bonus
from Employees
```

```sql
SELECT emp_id, name, COALESCE(salary, bonus) AS earnings
FROM Employees;
```

```sql
SELECT emp_id, name, COALESCE(salary, bonus,0) AS earnings
FROM Employees;
```

```sql
SELECT order_id, customer_name,
       COALESCE(amount, discount, 0) AS payment
FROM Orders;
```

```sql
SELECT emp_id, name, NULLIF(salary, 0) AS salary
FROM Employees;
```

```sql
select order_id,customer_name ,nullif (discount,0) as discount
from Orders
```

```sql
SELECT amount / NULLIF(discount, 0) as discount
FROM Orders;
```

```sql
select order_id,customer_name,nullif(coupon_code,'disc 10') as coupon_code
from Orders
```

```sql
SELECT name,
       COALESCE(salary, 0) + COALESCE(bonus, 0) AS total_earnings
FROM Employees;
```
