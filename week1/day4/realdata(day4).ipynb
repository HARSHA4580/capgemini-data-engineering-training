Create a unified email mapping table linking both emails to a single student_id.

```sql
Select regno ,personal_mail_id , college_domain_mailid from main_data
```

```sql
create or replace table mail_map_table as
select regno as student_id , college_domain_mailid as email from main_data
Union 
select regno  , personal_mail_id  from main_data
```

```sql
select* from mail_map_table
```

```sql
Select * from main_data a where not exists(select 1 from mail_map_table m join task1_table t on m.email = t.email_address where m.student_id = a.regno)
```

```sql
SELECT DISTINCT
       m.student_id,
       t.email_address,
       t.timestamp

FROM task2_table t

JOIN mail_map_table m
ON LOWER(t.email_address) = LOWER(m.email)

ORDER BY m.student_id;
```

valid submission


```sql
SELECT DISTINCT md.*

FROM task2_table t

INNER JOIN mail_map_table m
ON LOWER(t.email_address) = LOWER(m.email)

INNER JOIN main_data md
ON md.regno = m.student_id;
```

invalid submission


```sql
SELECT *

FROM (

        SELECT
               m.student_id,
               t.email_address,
               t.timestamp,

               ROW_NUMBER() OVER (
                       PARTITION BY m.student_id
                       ORDER BY t.timestamp
               ) AS rn

        FROM task2_table t

        INNER JOIN mail_map_table m
        ON LOWER(t.email_address) = LOWER(m.email)

) t

WHERE rn > 1;
```

```sql
SELECT
       student_id,
       email_address,
       timestamp,

       CASE
            WHEN rn = 1 THEN 'Valid Submission'
            ELSE 'Duplicate'
       END AS submission_status

FROM (

        SELECT
               m.student_id,
               t.email_address,
               t.timestamp,

               ROW_NUMBER() OVER (
                       PARTITION BY m.student_id
                       ORDER BY t.timestamp
               ) AS rn

        FROM task2_table t

        INNER JOIN mail_map_table m
        ON LOWER(t.email_address) = LOWER(m.email)

) t

ORDER BY student_id, timestamp;
     
```

Keep first submission, mark others as duplicates

```sql
SELECT
       m.student_id,
       t.email_address,
       t.timestamp,

       CASE
            WHEN ROW_NUMBER() OVER (
                    PARTITION BY m.student_id
                    ORDER BY t.timestamp
                 ) = 1
            THEN 'Submitted'
            ELSE 'Duplicate'
       END AS submission_status

FROM task2_table t

INNER JOIN mail_map_table m
ON LOWER(t.email_address) = LOWER(m.email)

ORDER BY m.student_id, t.timestamp;
```

GROUP BY Approach Vs Window Function Approach

```sql
SELECT
       m.student_id,
       COUNT(*) AS submission_count

FROM task2_table t
JOIN mail_map_table m
ON LOWER(t.email_address) = LOWER(m.email)

GROUP BY m.student_id;
```

submission count

```sql
SELECT
       m.student_id,
       COUNT(*) AS submission_count

FROM task2_table t

INNER JOIN mail_map_table m
ON LOWER(t.email_address) = LOWER(m.email)

GROUP BY m.student_id

ORDER BY submission_count DESC;
```

```sql
SELECT
       m.student_id,
       COUNT(DISTINCT LOWER(t.email_address)) AS email_used_count

FROM task1_table t

JOIN mail_map_table m
ON LOWER(t.email_address) = LOWER(m.email)

GROUP BY m.student_id

HAVING COUNT(DISTINCT LOWER(t.email_address)) = 2;
```

```sql
WITH valid_submissions AS (

    SELECT
           m.student_id,
           t.email_address,
           t.timestamp,

           ROW_NUMBER() OVER (
               PARTITION BY m.student_id
               ORDER BY t.timestamp
           ) AS rn

    FROM task2_table t

    JOIN mail_map_table m
    ON LOWER(t.email_address) = LOWER(m.email)

),

classified_valid AS (

    SELECT
           student_id,
           email_address,
           timestamp,

           CASE
                WHEN rn = 1 THEN 'Submitted'
                ELSE 'Duplicate'
           END AS final_status

    FROM valid_submissions
),

invalid_submissions AS (

    SELECT
           NULL AS student_id,
           t.email_address,
           t.timestamp,
           'Invalid' AS final_status

    FROM task2_table t

    LEFT JOIN mail_map_table m
    ON LOWER(t.email_address) = LOWER(m.email)

    WHERE m.student_id IS NULL
),

not_submitted AS (

    SELECT
           md.regno AS student_id,
           NULL AS email_address,
           NULL AS timestamp,
           'Not Submitted' AS final_status

    FROM main_data md

    WHERE NOT EXISTS (

        SELECT 1
        FROM mail_map_table m
        JOIN task2_table t
        ON LOWER(m.email) = LOWER(t.email_address)

        WHERE m.student_id = md.regno
    )
)

SELECT * FROM classified_valid

UNION ALL

SELECT * FROM invalid_submissions

UNION ALL

SELECT * FROM not_submitted;
```
